# SWDA

```python
import pandas as pd
import os
```

```python
trans = ['train_set_ru.csv', 
         'train_set_from_67140_ru.csv', 
         'train_set_from_104105_ru.csv',
         'train_set_from_129175_ru.csv',
         'train_set_from_129491_ru.csv',
         'train_set_from_129815_ru.csv',
         'train_set_from_129968_ru.csv',
         'train_set_from_136962_ru.csv',
         'train_set_from_137660_ru.csv',
         'train_set_from_143206_ru.csv',
         'train_set_from_168021_ru.csv',
         'train_set_from_178682_ru.csv']
```

```python
DIR_NAME = 'swda/train'
files = os.listdir(DIR_NAME)

data_train_ru = pd.DataFrame()
for item in trans:
    print(item)
    df_i = pd.read_csv(f'{DIR_NAME}/{item}')
    data_train_ru = pd.concat([data_train_ru, df_i])
print()
for item in files:
#     print(item)
    if item.endswith('.csv'):
        print(item)
print()

data_train = pd.DataFrame()
orig = []
for item in files:
#     print(item)
    if item.endswith('.csv') and not item.endswith('ru.csv'):
        print(item)
        orig.append(item)
        df_i = pd.read_csv(f'{DIR_NAME}/{item}')
        data_train = pd.concat([data_train, df_i])
```

```python
data_train_ru
```

```python
data_train_ru.to_csv('all_train_set_ru.csv', index=False)
```
