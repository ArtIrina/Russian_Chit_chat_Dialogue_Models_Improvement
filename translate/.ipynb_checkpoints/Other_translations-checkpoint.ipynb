Эта модель предсказывает вероятность того, что входное предложение приемлемо (грамматически верно) в русском языке. Вместо модели ruRoBERTa-large-rucola можно обучить свою на датасетах RuATD и/или RuCoLA

```python
from transformers import pipeline

classifier = pipeline("text-classification", model = "RussianNLP/ruRoBERTa-large-rucola", max_length=512, truncation=True)
```

```python
def check(test):
    ruRoBERTa = []
        
    for q in tqdm(test['ru']):
        ruRoBERTa.append(classifier(q)[0])
    
    ruRoBERTa = pd.DataFrame(ruRoBERTa)
    label_ruRoBERTa = ruRoBERTa['label']
    score_ruRoBERTa = ruRoBERTa['score']
    
    test.loc[:, 'label'] = label_ruRoBERTa
    test.loc[:, 'score'] = score_ruRoBERTa
    test
    
    # Сортируем по score по возрастанию
    ranked = test.sort_values(by='score')
    ranked.reset_index(inplace=True)
    
    # посмотрим на записи, переводы которых модель посчитала грамматически неверными
    display(ranked[ranked['label'] == 'LABEL_0']) 
    
    print('Процент записей, в лейблах которых, модель уверена менее, чем на 90%:', len(ranked[ranked['score'] <= 0.90]) / len(ranked) * 100)
    display(ranked[ranked['score'] <= 0.90])
    
    display(ranked[ranked['score'] > 0.90])
```

### **[0.75 балла]** Сравните качество перевода используя метрики BLEU и BLEURT. Для этого воспользуйтесь библиотекой [evaluate](https://github.com/huggingface/evaluate). Опишите принцип подсчета этих метрик, чем они отличаются, какие имеются преимущества и недостатки у каждой из этих метрик?

```python
from tqdm import tqdm
```

```python
pip install git+https://github.com/google-research/bleurt.git
```

```python
! pip install evaluate

import evaluate
bleu = evaluate.load("bleu")
bleurt = evaluate.load("bleurt")
```

```python
def evaluate_b(predict, refer, b):
  res_b = []
  for pred, ref in zip(predict, refer):
    res_b.append(b.compute(predictions=[pred], references=[ref]))
  
  ans_b = 0  
  for s in tqdm(res_b):
    if b == bleu:
      ans_b += s['bleu']
    if b == bleurt:
      ans_b += sum(s['scores'])/len(s['scores'])
  ans_b /= len(res_b)
  
  return ans_b
```

```python
print("BLEU на easyNMT:", evaluate_b(df['ru_to_en'], df['en'], bleu))
print("BLEU на Google:", evaluate_b(df['ru_to_en_google'], df['en'], bleu))
print("BLEURT на easyNMT:", evaluate_b(df['ru_to_en'], df['en'], bleurt))
print("BLEURT на Google:", evaluate_b(df['ru_to_en_google'], df['en'], bleurt))
```

Качество получилось ужасным во всех случаях

BLEU - это произведение штрафа за краткость (BP) и MEAN_BLEU.
MEAN_BLEU  является средним геометрическим значений (количество слов из потенциального перевода, содержащихся в эталонном/ых переводе/ах, деленное на количество слов в этом потенциальном переводе) для n-грамм, где n от 1 до 4.

При маленьком n - учитываем смысл предложения, при большом - важна правильная структура предложения.

Если наш перевод длиннее оригинала, то BP равен 1. Иначе, BP - это убывающая экспонента (чем больше разница длин переводов, тем ниже экспонента). При этом BP рассчитывается по всему корпусу, а не по отдельным переводам, что сглаживает штрафы за короткие переводы.

BLEURT - внутри этой метрики заложен BERT, синтетические данные и аннотации.
Считается, что эта метрика выдает более стабильные оценки, чем BLEU. Метрика позволяет работать с перефразированными переводами, учитывает смысл.Screenshot from 2022-11-25 16-24-50.png

То есть метрики BLEU и BLEURT совершенно разные

(https://neurohive.io/ru/novosti/bleurt-metrika-dlya-ocenki-modelej-dlya-generacii-teksta/, https://medium.com/nlplanet/two-minutes-nlp-learn-the-bleu-metric-by-examples-df015ca73a86)

```python
from statistics import mean

def eval_mean(res_b):
  ans_b = 0  
  for s in tqdm(res_b):
      ans_b += s['bleu']
  ans_b /= len(res_b)

  return ans_b
```

```python
# def evaluate_b_double(predict, en, refer, b):
predict = df_20k_elongated_sorted['diff_beam_size'] #
en = df_20k_elongated_sorted['en'] # 
refer = df_20k_elongated_sorted['ru'] #
b = bleu #


res_b = []
cur_res = []

prev_en = en[0]
prev_ru = refer[1]

for cur_pred, cur_en, cur_ref in tqdm(zip(predict, en, refer)):

  if prev_en == cur_en and prev_ru == cur_ref:
    cur_res.append(b.compute(predictions=[cur_pred], references=[cur_ref])) 
  else:
    res_b.append(eval_mean(cur_res)) # усредняем качество для одинаковых en и ru
    cur_res = []
    cur_res.append(b.compute(predictions=[cur_pred], references=[cur_ref]))
      
res_b.append(eval_mean(cur_res))

ans_b = sum(res_b) / len(res_b) # усредняем качество перевода
  
print(ans_b)
  # return ans_b
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
def get_label(inputs, tokenizer, model):
  with torch.no_grad():
      logits = model(**inputs).logits

  predicted_class_id = logits.argmax().item()
  return model.config.id2label, model.config.id2label[predicted_class_id]

def get_score(inputs, tokenizer, model, for_length):  
  num_labels = len(for_length)
  model = AutoModelForSequenceClassification.from_pretrained("RussianNLP/ruRoBERTa-large-rucola", num_labels=num_labels)

  labels = torch.tensor([1])
  loss = model(**inputs, labels=labels).loss
  return round(loss.item(), 2)
```

```python
tokenizer = AutoTokenizer.from_pretrained("RussianNLP/ruRoBERTa-large-rucola")

model = AutoModelForSequenceClassification.from_pretrained("RussianNLP/ruRoBERTa-large-rucola")

ruRoBERTa_label = []
ruRoBERTa_score = []

for q in tqdm(df_20k_elongated_sorted['diff_beam_size']):
  inputs = tokenizer(q, return_tensors="pt")
  for_length, l = get_label(inputs, tokenizer, model)
  ruRoBERTa_label.append(l)
  ruRoBERTa_score.append(get_score(inputs, tokenizer, model, for_length))
```

При запуске, как в найденном примере из документации, прогнозируется 60 часов на процедуру ранжирования, так что я оставлю предыдущую реализацию. В ином случае, придется долго-долго ждать, после чего в массиве ruRoBERTa_label будут лежать все label подряд, а в ruRoBERTa_score - значения score. Далее повторить действия выше: добавить столбцы label и score к датафрейму, отсортировать по score по возрастанию, из дубликатов оставить последний (с наибольшим score)
