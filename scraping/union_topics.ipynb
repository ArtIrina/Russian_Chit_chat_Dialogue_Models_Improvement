```python
from sklearn.model_selection import train_test_split
```

```python
sport_train = pd.read_csv('sport_train.csv')
sport_train
```

```python
sport_test = pd.read_csv('sport_test.csv')
sport_test
```

```python
sport_train, sport_val = train_test_split(sport_train, test_size=0.2, random_state=42)
```

```python
sport_train
```

```python
sport_val
```

```python
sport_train.loc[:, 'topic'] = ['Sports'] * len(sport_train)
sport_test.loc[:, 'topic'] = ['Sports'] * len(sport_test)
sport_val.loc[:, 'topic'] = ['Sports'] * len(sport_val)
```

```python
topic_data = topic_data.append(sport_train[['text', 'topic']])
topic_data = topic_data.append(sport_test[['text', 'topic']])
topic_data = topic_data.append(sport_val[['text', 'topic']])
topic_data
```

```python
topic_data.to_csv('topics_classification.csv', index=False)
```
