```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

df = pd.DataFrame(columns=['text'])
df
```

```python
url = 'https://fantlab.ru/forum/forum6page1'
```

```python
response = requests.get(url)
print(response)
```

```python
bs = BeautifulSoup(response.text,"lxml")
print(bs)
```

```python
temp = bs.find_all("font", "v9b")
```

```python
print(temp)
```

```python
print(temp[0].text)
```

```python
len(temp)
```

```python
for i in range(len(temp)):
    df.loc[ len(df.index )] = [temp[i].text]
```
