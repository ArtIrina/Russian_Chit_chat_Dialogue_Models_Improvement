```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import requests
import json

def get_html(url):
    response = requests.get(url)
    return response.text

def get_all_links(html):
    soup = BeautifulSoup(html, features="html.parser")
    tds = soup.find_all('td', 'windowbg')
    links = []
    for td in tds:
        if td.a != None and not td.a.has_attr('class'):
            if len(td.span) > 0:
                l = td.span.find_all('a')[-1]['href']
                links.append(l)
            else:
                l = td.a['href']
                links.append(l)
    return links

def get_page_data(html):
    soup = BeautifulSoup(urlopen(html), features="html.parser")

    # user
    tds = soup.find_all('div', 'displaycontainer')
    users = []
    for td in tds:
        if td.span != None and not td.span.has_attr('class'):
            u = td.span.find('b').text
            users.append(u)
#     print(users)

    # time
    tds = soup.find_all('div', 'displaycontainer')#'td', {'align':"left", 'class':"windowbg", 'valign':"top", 'width':"140"})
    times = []
    for td in tds:
        ti = td.find_all('span', 'class'=='small')
        for t in ti:
            if t.span != None and not t.span.has_attr('style'):
                time = t.span.text
                times.append(time)
#     print(times)


    # text (надо вычесть цитату)
    tds = soup.find_all("div", "message")
    tex = []
    for td in tds:   
        tex.append(td.text)
#     print(tex)
    

    # quote
    tds = soup.find_all("div", "message")
    temp = []
    for td in tds:   
        t = td.find("div", "quote")
        if t != None:
            temp.append(t.text)
        else:
            temp.append('')
#     print(temp)

    data = ({'user': users,                    
             'time': times,      
             'text': tex,            
             'quote': temp,          
             'topic': ['Books&Literature'] * len(users)})          
    return users, times, tex, temp, ['Books&Literature'] * len(users)#data

def page_content(all_links, table):
        print(all_links)
        us, ti, te, tem, l = get_page_data(all_links)
#         table.append(data)
        for i in range(len(us)):
            if i > 5:
                break
            users.append(us[i])
            times.append(ti[i])
            tex.append(te[i])
            temp.append(tem[i])
            labels.append(l[i])

table = pd.DataFrame(columns={'users', 'times', 'text', 'quote', 'topic'})
users = []
times = []
tex = []
temp = []
labels = []

url = 'http://publ.lib.ru/cgi/forum/YaBB.pl?board=talk'
all_links = get_all_links(get_html(url))
for i in range(len(all_links)):
    page_content(all_links[i], table)
    print(i)
    
table.loc[:, 'users'] = users
table.loc[:, 'times'] = times
table.loc[:, 'text'] = tex
table.loc[:, 'quote'] = temp
table.loc[:, 'topic'] = labels
```
